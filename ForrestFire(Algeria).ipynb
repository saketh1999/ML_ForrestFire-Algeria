# **Algerian Forest Fires Dataset**

**Data Set Information:**

The dataset includes 244 instances that regroup a data of two regions of Algeria,namely the Bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria.

122 instances for each region.

The period from June 2012 to September 2012. The dataset includes 11 attribues and 1 output attribue (class) The 244 instances have been classified into fire(138 classes) and not fire (106 classes) classes.

Attribute Information:

1. Date : (DD/MM/YYYY) Day, month ('june' to 'september'), year (2012) Weather data observations
2. Temp : temperature noon (temperature max) in Celsius degrees: 22 to 42
3. RH : Relative Humidity in %: 21 to 90
4. Ws :Wind speed in km/h: 6 to 29
5. Rain: total day in mm: 0 to 16.8 FWI Components
6. Fine Fuel Moisture Code (FFMC) index from the FWI system: 28.6 to 92.5
7. Duff Moisture Code (DMC) index from the FWI system: 1.1 to 65.9
8. Drought Code (DC) index from the FWI system: 7 to 220.4
9. Initial Spread Index (ISI) index from the FWI system: 0 to 18.5
10. Buildup Index (BUI) index from the FWI system: 1.1 to 68
11. Fire Weather Index (FWI) Index: 0 to 31.1
12. Classes: two classes, namely Fire and not Fire

# **Data Cleaning**

```python
dataset.isnull().sum() #This helps us find the sum of the number of null values present in each column
```

```python
dataset[dataset.isnull().any(axis=1)]
#  1. dataset.isnull()

# This part identifies missing values in the dataset.
# In pandas, missing values are typically represented by NaN (Not a Number) or None.
# The .isnull() method creates a DataFrame of the same shape as the original dataset, but with True for missing values and False for valid values.
# 2. .any()

# This method is applied to the DataFrame returned by .isnull().
# The .any() method checks if there are any True values (i.e., missing values) present in any column.
# In the context of a DataFrame, .any() by default checks along the columns (axis=0).
```

The Dataset is converted into two sets based on Regions from 122th index , we can make a new column based on the Region

1 : "Bejaia Region Dataset" - > the new column called region will represent this region as 0

2 : "Sidi-Bel Abbes Region Dataset" - > the new column called region will represent this region as 1

Add new column with region

```python
df = dataset
df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1
#More on Loc function - https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.loc.html#pandas.DataFrame.loc
df[['Region']] = df[['Region']].astype(int)
df.info()
```

```python
##Removing the null values
df = df.dropna().reset_index(drop=True)
df.head()
# All Null Values removed
df.isnull().sum()
```

```python
df.iloc[[122]]
#But the row is still present in the df
```

```python
df = df.drop(122).reset_index(drop=True)

# df.drop(122): This attempts to remove the row at index 122 from the DataFrame df. However, if there's no row with that specific index, it raises a KeyError.

# reset_index(drop=True): This part comes into play even if the drop(122) fails. Here's what reset_index does:

# Renumbering the Index: By default, reset_index converts the existing index (which could be integers, labels, or custom) into a new default integer index starting from 0 and incrementing by 1 for each row. This ensures a continuous and predictable index for the DataFrame.

# Optional drop=True: The argument drop=True tells reset_index to not include the old index as a new column in the DataFrame. This is because if drop(122) didn't raise an error, the old index might contain the dropped row's label, which can be confusing.



#We are completely droping the row 122
df.iloc[[122]] #Now we can see that 122 row is properly pointing to the data we need
```

```python
df.columns
```

```python
#Fixing the spaces in column names
df.columns = df.columns.str.strip()

# The code df.columns.str.strip() is used in Pandas to remove leading and trailing whitespaces from the column names of a DataFrame. Here's a breakdown of the code:
# df: This represents a Pandas DataFrame object.
# .columns: This refers to the column labels (names) of the DataFrame.
# .str.strip(): This part applies the strip() function to each column name as a string.
# The str accessor is used to indicate that we're working with strings. The strip() function itself removes any whitespace characters (including spaces, tabs, and newlines) from the beginning and end of the string.

df.columns
```

# **Changing the required Columns as integer data type**

```python
df[['day', 'month', 'year', 'Temperature', 'RH','Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH','Ws']].astype(int)
df.info()
```

```python
#Changing the remaining columns to float data type
objects = [features for features in df.columns if df[features].dtype=='O']

for i in objects:
  if i!="Classes":
    df[[i]]=df[[i]].astype(float)

df.info()
```
